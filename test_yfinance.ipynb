```python
key_metrics = income_statement.loc[['Total Revenue', 'Gross Profit', 'Operating Income', 'Net Income', 'EBITDA']]
print(key_metrics)
```

```python
key_metrics_clean = key_metrics.iloc[:, :4] / 1e9
key_metrics_clean = key_metrics_clean.round(2)
print(key_metrics_clean)
```

```python

search_results = yf.Search("Apple")
print(search_results.quotes)
```

```python
results = yf.Search("Apple").quotes

for r in results:
    print(r.get('longname'), "—", r.get('symbol'), "—", r.get('exchDisp'))
```

```python
us_results = [r for r in results if r.get('exchDisp') in ['NASDAQ', 'NYSE']]

for r in us_results:
    print(r.get('longname'), "—", r.get('symbol'), "—", r.get('exchDisp'))
```
